```python
import pandas as pd
import unicodedata
```

```python
def strip_accents(s):
    return ''.join(c for c in unicodedata.normalize('NFD', s) if unicodedata.category(c) != 'Mn')
```

```python
df = pd.read_excel('enunciados.xlsx')
df.columns = [strip_accents(column) for column in df.columns]
df.to_csv('enunciados.csv')
```

```python
df
```

```python
agrupados = df.groupby(['Area','Tema', 'Subtema']).count().sort_values(by=['Enunciado'], ascending=False)
```

```python
agrupados.to_excel('enunciados-agrupados.xls')
```
